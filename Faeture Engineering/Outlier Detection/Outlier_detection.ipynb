# <center>Outlier detection and removal using percentile</center>

```python
import pandas as pd

df = pd.read_csv("height.csv")
```

```python
df.head()
```

### Let's check maximum thresold

```python
max_thresold = df['height'].quantile(0.95)
max_thresold
```

```python
df[df['height']>max_thresold]
```

this data point is showing the height of imran is 14.5 and that's not possible

### Check the minimum thresold percentile

```python
min_thresold = df['height'].quantile(0.05)
min_thresold
```

```python
df[df['height']<min_thresold]
```

## Removing this Outliers

```python
df[(df['height']<max_thresold) & (df['height']>min_thresold)]
```

```python
df1= pd.read_csv('Bengaluru_House_Data.csv')
```

```python
def convert_sqft_to_num(x):
    tokens = x.split('-')
    if len(tokens) == 2:
        return (float(tokens[0])+float(tokens[1]))/2
    try:
        return float(x)
    except:
        return None
```

```python
df2 = df1.copy()
df2.total_sqft = df2.total_sqft.apply(convert_sqft_to_num)
df1 = df1[df1.total_sqft.notnull()]
df1.head(2)
```

```python
df2.loc[30]
```

#### Add new feature called price per square feet

```python
df3 = df2.copy()
df3['price_per_sqft'] = df2['price']*100000/df2['total_sqft']
df3.head()
```

## Now check the thresold

```python
minimum_thresold, maximum_thresold = df3.price_per_sqft.quantile([0.001, 0.999])
```

```python
df3[df3.price_per_sqft < minimum_thresold]
```

you can see the price per sqfeet is between 8 to 10 rs and that is not possible

```python
df3[df3.price_per_sqft>maximum_thresold]
```

Now let's remove this outliers

```python
df4 = df3[(df3.price_per_sqft<maximum_thresold) & (df3.price_per_sqft>minimum_thresold)]
```
